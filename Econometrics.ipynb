## goldfeld quandt test


```python
from scipy.stats import f
```

```python
def rss(x, y):
    model = sm.OLS(y, sm.add_constant(x))
    results = model.fit()

    y_pred = results.predict(sm.add_constant(x))
    RSS = ((y - y_pred) ** 2).sum()

    return RSS
```

```python
def goldfeld_quandt(x, y, split):
    """
    Goldfeld-Quandt test for heteroscedasticity.

    Parameters:
    x (array-like): Independent variable.
    y (array-like): Dependent variable.
    split (float): Proportion of data to be used in the middle of the range.

    Returns:
    F (float): Test statistic.
    p_value (float): Two-tailed p-value.
    """
    n = len(x)
    k = int(np.floor(n * split))

    x1 = x[:k]
    y1 = y[:k]
    x2 = x[-k:]
    y2 = y[-k:]

    res1 = rss(x1, y1)
    res2 = rss(x2, y2)

    f_stat = res2 / res1

    return f_stat
```
